```python
import os

# 하위 폴더의 경로를 포함한 모든 폴더 리스트를 가져옵니다.
def get_subdirectory_list(parent_directory):
    subdirectory_list = []
    for root, dirs, files in os.walk(parent_directory):
        for name in dirs:
            subdirectory_list.append(os.path.join(root, name))
    return subdirectory_list

# 해당 폴더 내의 모든 하위 폴더의 이름 리스트를 가져옵니다.
def get_subdirectory_name_list(parent_directory):
    subdirectory_list = get_subdirectory_list(parent_directory)
    subdirectory_name_list = []
    for directory in subdirectory_list:
        subdirectory_name_list.append(os.path.basename(directory))
    return subdirectory_name_list
```

```python
searchlist=get_subdirectory_list('미니프로젝트')
```

```python
#영양소사이트크롤링

import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re




htmls = []


for i in sea:
    url = f'https://www.fatsecret.kr/%EC%B9%BC%EB%A1%9C%EB%A6%AC-%EC%98%81%EC%96%91%EC%86%8C/%EC%9D%BC%EB%B0%98%EB%AA%85/{i}'
    html = requests.get(url)
    htmls.append(html)
```

```python


#입력정보 불러오기 
result=[]
for html in htmls:
    soup=bs(html.text,'html.parser')
    res=[]
    for i in soup.find_all('div','nutrition_facts international'):
        name=soup.find('title').text.split('안')[0]
        cal=i.find('div',"nutrient right tRight").text
        tan=i.find_all('div','nutrient black right tRight')[1].text
        dan=i.find_all('div','nutrient black right tRight')[2].text
        ji=i.find_all('div','nutrient black right tRight')[3].text
        tmp={
        '음식이름':name,
        '칼로리':cal,
        '지방':ji,
        '단백질':dan,
            
        '탄수화물':tan          
        }
        res.append(tmp)
    df=pd.DataFrame(res)
    result.append(df)
        
```

```python
#모은 자료 콘캣하기 
df_res=pd.concat(result,ignore_index=True)
display(df_res)
```

```python
df_res.to_csv('KFood.csv')  #데이터프레임 csv로 저장 
```
