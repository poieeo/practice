```python
import matplotlib.pyplot as plt
import numpy as np


def handle_image(x,option=1):                #핸들이미지 함수 식으로 설정하는것 
    #이미지 읽어오기
    imgfile=x
    if option==1:
        img = cv2.imread(imgfile, cv2.IMREAD_COLOR) #사진을 컬러로 읽어오기 
        caption='color' 
    
    else:
        
        img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE) #사진을 그레이컬러 
        caption='gray' 
    caption='image info'+caption+'/filename:'+x
    cv2.imshow(caption,img)  #창하나 띄워주세요 #제목명으로 이미지가 나옴 image 
    cv2.waitKey(0) # 키보드 입력을 대기하는 함수, milisecond값을 넣으면 해당 시간동안 대기, 0인경우 무한으로 대기
    
    plt.imshow(img)
handle_image('Golde33443.jpg',option=0)        #handle_image(option=1)
```

```python
import cv2

capture = cv2.VideoCapture(0)
capture.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

while cv2.waitKey(33) < 0:
    ret, frame = capture.read()
    cv2.imshow("VideoFrame", frame)

capture.release()
cv2.destroyAllWindows()
```

```python
import cv2

x='Golde33443.jpg'
    

src = cv2.imread(x, cv2.IMREAD_COLOR)
dst = cv2.flip(src, 0)

cv2.imshow('src', src)  #정방향 
cv2.imshow("dst", dst)  #그것을 거꾸로 
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread(x, cv2.IMREAD_COLOR)

height, width, channel = src.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
dst = cv2.warpAffine(src, matrix, (width, height))

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread(x, cv2.IMREAD_COLOR)
height, width, channel = src.shape

dst = cv2.pyrUp(src, dstsize=(width * 2, height * 2), borderType=cv2.BORDER_DEFAULT)
dst2 = cv2.pyrDown(src)

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.imshow("dst2", dst2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread(x, cv2.IMREAD_COLOR)

dst = cv2.resize(src, dsize=(640, 480), interpolation=cv2.INTER_AREA)
dst2 = cv2.resize(src, dsize=(0, 0), fx=0.3, fy=0.7, interpolation=cv2.INTER_LINEAR)

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.imshow("dst2", dst2)
cv2.waitKey()
cv2.destroyAllWindows()
```
