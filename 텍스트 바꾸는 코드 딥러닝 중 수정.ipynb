###  트레인 라벨,테스트 라벨, 바리드 라벨 변환하기


### readline() - 파일의 한 줄을 가져와 문자열로 반환합니다. 파일 포인터는 그 다음줄로 이동합니다.

### readlines() - 파일 내용 전체를 가져와 리스트로 반환합니다. 각 줄은 문자열 형태로 리스트의 요소로 저장됩니다.

#### 예를들어 5줄짜리 파일을 readlines() 로 읽게 되면 문자열 5개를 요소로 갖는 리스트가 반환됩니다

```python


# 폴더안의 파일읽어서 첫글자를 내가 지정한 내용으로 변경하기
from glob import glob
fileList=glob('./testlabel/*.txt')   #텍스트파일의 리스트들을 가지고오기 파일리스트로 
#print(fileList)

for file in fileList:               #하나씩 하니씩 불러오기 
    with open(file, "r") as f:      #오픈파일하기 리드 불러오기 라인스는 리드라인스로 리스트 처럼 만들기 
        lines = f.readlines()
        
    with open(file, "w") as f:   # 다시 쓰기 w는 다시 써주기 위에 r, 즉 리드를 통하여 다시 불러온것을 부른다 이것을위해 위에 r이필요함
        for line in lines:
            a=line.split()   #라인 스플릿하면 리스트로 전환됨 그것을 a로 치환
            a[0]=2  #a의 0값을 1로전환한다
            
            line=str(a[0])+line[1:]    # a[0]을 1로바꾸고 그것을 str로 바꾸고 라인에서 0 자리뺀 1부터 끝까지간다

            f.write(f'{line}')     #라인을 f포맷하여 f write를 하여 다시쓴다 
            

            
    print(file + '------작업 end')
   
```
